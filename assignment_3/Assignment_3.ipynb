```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, precision_recall_curve, precision_score, recall_score, f1_score,average_precision_score, log_loss
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
url = 'https://raw.githubusercontent.com/Dhanush-adk/machine_learning/main/assignment_3/diabetes.csv'
df = pd.read_csv(url)

# Display the first 5 rows of the DataFrame
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
X = df.drop(columns=['Outcome'])
y = df['Outcome']
```

```python
X.shape, y.shape
```

```python
# We specify this so that the train and test data set always have the same rows, respectively
np.random.seed(9)
X_train, X_test, y_train, y_test = train_test_split(X,
                                                   y,
                                                   train_size = 0.8,
                                                   test_size = 0.20,
                                                   random_state = 100)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train.head()
```

```python
scale= StandardScaler()
X_train_scaled = scale.fit_transform(X_train) # standardization of dependent variables
X_train_scaled.shape
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(X_train_scaled, y_train)
```

```python
model.coef_
```

```python
model.score(X_train_scaled, y_train)
```

```python
X_test_scaled = scale.transform(X_test)
y_pred = model.predict(X_test_scaled)
```

```python
cnf_matrix = confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix = cnf_matrix,
                             display_labels = model.classes_)
disp.plot()
```

```python
accuracy_score(y_test, y_pred), precision_score(y_test, y_pred), recall_score(y_test, y_pred), f1_score(y_test, y_pred)
```

```python
probs = model.predict_proba(X_test_scaled)
precision, recall,  thersholds = precision_recall_curve(y_test, probs[:,1])
```

```python
plt.figure(figsize=(8, 6))
plt.plot(precision, recall, marker='.')
plt.xlabel('Precision')
plt.ylabel('Recall')
plt.title('Precision-Recall Curve')
plt.grid(True)
plt.show()
```

```python
y_pred_prob = model.predict_proba(X_test)
log_loss(y_test, y_pred_prob)
```

```python
url_cancer = 'https://raw.githubusercontent.com/Dhanush-adk/machine_learning/main/assignment_3/cancer.csv'
data = pd.read_csv(url_cancer)

# Display the first 5 rows of the DataFrame
data.head()
```

```python
data.info()
```

```python
data = data.drop(columns = ['id', 'Unnamed: 32'])
```

```python
data['diagnosis'].replace({'B': 0, 'M': 1}, inplace=True)
data.head()
```

```python
X = data.drop(columns=['diagnosis'])
y = data['diagnosis']
```

```python
# We specify this so that the train and test data set always have the same rows, respectively
np.random.seed(9)
X_train, X_test, y_train, y_test = train_test_split(X,
                                                   y,
                                                   train_size = 0.8,
                                                   test_size = 0.20,
                                                   random_state = 100)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
scale= StandardScaler()
X_train_scaled = scale.fit_transform(X_train) # standardization of dependent variables
X_train_scaled.shape
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(X_train_scaled, y_train)
```

```python
model.score(X_train_scaled, y_train)
```

```python
X_test_scaled = scale.transform(X_test)
y_pred = model.predict(X_test_scaled)
```

```python
accuracy_score(y_test, y_pred)
```

```python
cnf_matrix = confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred), precision_score(y_test, y_pred), recall_score(y_test, y_pred), f1_score(y_test, y_pred)
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix = cnf_matrix,
                             display_labels = ['B', 'M'])
disp.plot()
```

```python
probs = model.predict_proba(X_test_scaled)
precision, recall,  thersholds = precision_recall_curve(y_test, probs[:,1])
```

```python
plt.figure(figsize=(8, 6))
plt.plot(precision, recall, marker='.')
plt.xlabel('Precision')
plt.ylabel('Recall')
plt.title('Precision-Recall Curve')
plt.grid(True)
plt.show()
```

```python
y_pred_prob = model.predict_proba(X_test)
log_loss(y_test, y_pred_prob)
```
